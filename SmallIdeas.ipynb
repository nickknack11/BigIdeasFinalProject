```python
import pandas as pd
df = pd.read_csv("NeighborhoodParks.csv")
num_parks = (df["neighborhood"].value_counts())
num_parks.plot(kind = "bar", figsize = (28,12))
```

```python
#puts neighborhoods as the key and how many parks they have as the value in a dictionary
neighborhoods_dictionary = {}
for index, rows in df.iterrows():
    if(rows["neighborhood"] in neighborhoods_dictionary):
        neighborhoods_dictionary[rows["neighborhood"]] += 1
    else:
        neighborhoods_dictionary[rows["neighborhood"]] = 1
print(neighborhoods_dictionary)
```
